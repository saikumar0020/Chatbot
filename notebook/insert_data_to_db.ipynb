```python
from dateutil import parser

def convert_to_standard_format(date_str):
    """
    Convert a date string to 'YYYY-MM-DD HH:MM' format.
    
    Args:
        date_str (str): Input date string in any common format.

    Returns:
        str: Formatted date string in 'YYYY-MM-DD HH:MM'.
    """
    try:
        dt = parser.parse(date_str)
        return dt.strftime('%Y-%m-%d %H:%M')
    except Exception as e:
        return None
```

```python
import psycopg2

# Connect to PostgreSQL
conn = psycopg2.connect(
    dbname="testdb",
    user="postgres",
    password="mysecretpassword",
    host="localhost",
    port="5432"
)

cur = conn.cursor()

# Raw SQL to create table
create_table_query = """
CREATE TABLE IF NOT EXISTS invoices (
    document_id VARCHAR PRIMARY KEY,
    doc_received_date TIMESTAMP,
    invoice_status VARCHAR,
    invoice_type VARCHAR,
    invoice_number VARCHAR,
    invoice_date DATE,
    po_number VARCHAR,
    currency VARCHAR,
    tax_amount NUMERIC,
    invoice_amount NUMERIC,
    supplier_name TEXT,
    bill_to TEXT
);
"""

# Execute and commit
cur.execute(create_table_query)
conn.commit()

cur.close()
conn.close()

print("Table created using raw SQL.")
```

```python
from sqlalchemy import create_engine

# Create PostgreSQL engine
engine = create_engine("postgresql://postgres:mysecretpassword@localhost:5432/testdb")
```
